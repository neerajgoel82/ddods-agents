```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from bkmeans import BKMeans as BreathingKMeans

sns.set()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/gittar/breathing-k-means/master/data/D31.csv", 
                 names = ["x1", "x2"])
```

```python
# number of centroids
k=31

# get data as numpy array
X = df.values
```

```python
fig, ax = plt.subplots(figsize = (12, 12))

plt.scatter(X[:, 0], X[:, 1], c = '#00B9FF')
plt.title("Dataset", fontsize = 30, fontweight="bold")
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show();
```

```python
%%timeit

kmp=KMeans(n_clusters=k)
kmp.fit(X)
```

```python
%%timeit

bkm=BreathingKMeans(n_clusters=k)
bkm.fit(X)
```

```python
kmp=KMeans(n_clusters=k)
kmp.fit(X)

bkm=BreathingKMeans(n_clusters=k)
bkm.fit(X)

print(f"KMeans  SSE: {kmp.inertia_:.3E}")
print(f"BKMeans SSE: {bkm.inertia_:.3E} ({1-bkm.inertia_/kmp.inertia_:.2%} lower)")

fig, ax = plt.subplots(figsize = (9, 9))

plt.scatter(X[:, 0], X[:, 1], label='Data Point',c='#82E0AA')
plt.scatter(kmp.cluster_centers_[:, 0], kmp.cluster_centers_[:, 1], c='red', marker='o', label='Centroid', s = 100)
plt.title("KMeans++", fontsize = 40, fontweight="bold", pad = 15)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.legend()
plt.show()

fig, ax = plt.subplots(figsize = (9, 9))

plt.scatter(X[:, 0], X[:, 1], label='Data Point',c='#82E0AA')
plt.scatter(bkm.cluster_centers_[:, 0], bkm.cluster_centers_[:, 1], c='red', marker='o', label='Centroid', s = 100)
plt.title("Breathing KMeans", fontsize = 40, fontweight="bold", pad = 15)
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.legend()
plt.show()
```

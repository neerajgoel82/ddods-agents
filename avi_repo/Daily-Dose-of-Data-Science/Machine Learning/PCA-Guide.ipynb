```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
sns.set()
```

```python
x = np.linspace(1, 5, 100)
y = 2 * x + 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Dataset with Noise: y = 2x + 3')
plt.show()
```

```python
data = np.column_stack((x, y))
```

```python
pca = PCA(n_components=1)
pca.fit(data)
```

```python
# Project the data onto the first principal component
projected_data = pca.transform(data)

# Reconstruct the data using the first principal component
reconstructed_data = pca.inverse_transform(projected_data)

plt.scatter(x, y, label='Original Data')
plt.scatter(reconstructed_data[:, 0], reconstructed_data[:, 1], label='Reconstructed Data', alpha=0.7)
plt.xlabel('x')
plt.ylabel('y')
plt.title('PCA with 1 Component')
plt.legend()
plt.show()
```

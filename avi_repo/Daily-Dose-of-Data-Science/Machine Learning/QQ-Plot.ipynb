# Visualize The Performance Of Linear Regression With This Simple Plot

Post Link: [Substack](https://www.blog.dailydoseofds.com/p/visualize-the-performance-of-linear)

LinkedIn Post: [LinkedIn](https://www.linkedin.com/feed/update/urn:li:share:7070345486662049792/)

Twitter Post: [Twitter](https://twitter.com/_avichawla/status/1664579801665945601)

```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
def feature_func(x):
    return x**5
```

```python
# Generate data
points = 150
x = np.linspace(0, 5, points)
y = feature_func(x) + np.random.normal(0, 100, points)
```

```python
# Simple linear regression (1st column, 1st row)
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
y_pred = model.predict(X)

# QQ plot (1st column, 2nd row)
residuals = y - y_pred
```

```python
# Linear regression with feature (2nd column, 1st row)
X_feature = sm.add_constant(feature_func(x))
model_feature = sm.OLS(y, X_feature).fit()
y_pred_feature = model_feature.predict(X_feature)

# QQ Plot (2nd column, 2nd row)
residuals_feature = y - y_pred_feature
```

```python
colors = ['#fe7c73', '#2471A3', '#3498DB', '#27AE60', '#82E0AA', '#D35400', '#5D6D7E', '#E74C3C', '#21618C', '#B7950B', '#46C7C7', '#00B9FF', '#FF7051', "orange", "darkorange", "tomato", "coral", "limegreen", "lightsalmon"]
```

```python
# Plotting
fig, axs = plt.subplots(2,2, figsize=(12, 9))

# 1st column, 1st row: simple linear regression
axs[0, 0].scatter(x, y  , color=colors[2])
axs[0, 0].plot(x, y_pred, color=colors[12], lw=3 )
axs[0, 0].set_title('Regression Fit', fontsize = 20, fontweight = "bold", pad = 15)
axs[0, 0].set_xticklabels([])
axs[0, 0].set_yticklabels([])
axs[0, 0].set_ylabel("Model A",fontsize = 20, fontweight = "bold")

# 1st column, 2nd row: QQ Plot
sm.qqplot(residuals, line='s', ax=axs[0, 1])
axs[0, 1].set_title('Q-Q Plot', fontsize = 20, fontweight = "bold", pad = 20)
axs[0, 1].get_lines()[1].set_color(colors[12])
axs[0, 1].get_lines()[0].set_markerfacecolor(colors[2]) 
axs[0, 1].get_lines()[1].set_linewidth(3)
axs[0, 1].get_lines()[0].set_markeredgecolor(colors[2]) 

# 2nd column, 1st row: linear regression with feature
axs[1, 0].scatter(x, y, color=colors[2])
axs[1, 0].plot(x, y_pred_feature, color=colors[12], lw=3)
axs[1, 0].set_xticklabels([])
axs[1, 0].set_yticklabels([])
axs[1, 0].set_ylabel("Model B",fontsize = 20, fontweight = "bold")


# 2nd column, 2nd row: QQ Plot
sm.qqplot(residuals_feature, line='s', ax=axs[1, 1])
axs[1, 1].get_lines()[1].set_color(colors[12])
axs[1, 1].get_lines()[0].set_markerfacecolor(colors[2]) 
axs[1, 1].get_lines()[1].set_linewidth(3)
axs[1, 1].get_lines()[0].set_markeredgecolor(colors[2]) 

plt.tight_layout()
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()
```

```python
# Generate a skewed dataset using the gamma distribution
shape = 2.0
scale = 2.0
size = 1000

# Generate the skewed dataset
skewed_data = np.random.gamma(shape, scale, size)
```

```python
color_code = '#fe7c73'

colors = ['#fe7c73', '#2471A3', '#3498DB', '#27AE60', '#82E0AA', '#D35400', '#5D6D7E', '#E74C3C', '#21618C', '#B7950B', '#46C7C7', '#00B9FF', '#FF7051', "orange", "darkorange", "tomato", "coral", "limegreen", "lightsalmon"]
```

```python
plt.hist(skewed_data, bins=30, edgecolor='black', color = color_code)
plt.title('Original Data Distribution', fontsize = 20, fontweight = "bold", pad = 10)
plt.xticks([0,3,6,9,12,15]);
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
standardized_data = scaler.fit_transform(skewed_data.reshape(-1, 1))

# Plot the histogram of the standardized dataset
plt.hist(standardized_data, bins=30, edgecolor='black', color =color_code)

plt.title('Standardized Data Distribution', fontsize = 20, fontweight = "bold", pad = 10);
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(skewed_data.reshape(-1, 1))

# Plot the histogram of the scaled dataset
plt.hist(scaled_data, bins=30, edgecolor='black', color = colors[18])

plt.show()
```

```python
# Generate data for the three distributions
size = 1000

# Normal distribution
mean = 0
std = 1
normal_data = np.random.normal(mean, std, size)

# Uniform distribution
low = 0
high = 1
uniform_data = np.random.uniform(low, high, size)

# Skewed distribution (Gamma)
shape = 2.0
scale = 2.0
skewed_data = np.random.gamma(shape, scale, size)

# Create subplots for the first row (original distributions)
fig, axes = plt.subplot_mosaic("ABC;DEF", figsize = (15, 10))

# Plot the original distributions
axes["A"].hist(normal_data, bins=30, edgecolor='black', color =color_code)
axes["A"].set_title('Normal Distribution', fontsize = 20, fontweight = "bold")
axes["A"].set_ylabel('Before Standardization', fontsize = 20, fontweight = "bold")

axes["B"].hist(uniform_data, bins=30, edgecolor='black', color =color_code)
axes["B"].set_title('Uniform Distribution', fontsize = 20, fontweight = "bold")

axes["C"].hist(skewed_data, bins=30, edgecolor='black', color =color_code)
axes["C"].set_title('Skewed Distribution', fontsize = 20, fontweight = "bold")


scaler = StandardScaler()

normalized_normal = scaler.fit_transform(normal_data.reshape(-1, 1))
normalized_uniform = scaler.fit_transform(uniform_data.reshape(-1, 1))
normalized_skewed = scaler.fit_transform(skewed_data.reshape(-1, 1))

axes["D"].hist(normalized_normal, bins=30, edgecolor='black', color = colors[18])
axes["D"].set_ylabel('After Standardization', fontsize = 20, fontweight = "bold")

axes["E"].hist(normalized_uniform, bins=30, edgecolor='black', color = colors[18])

axes["F"].hist(normalized_skewed, bins=30, edgecolor='black', color = colors[18])
plt.show()
```

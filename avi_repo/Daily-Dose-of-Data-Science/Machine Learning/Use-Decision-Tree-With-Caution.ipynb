```python
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

import seaborn as sns
sns.set()
```

```python
c_list = ['#46C7C7', '#00B9FF', '#FF7051'][::-1]
plt.rcParams['font.family'] = ['Times New Roman', 'sans-serif']
```

```python
np.random.seed(123)
```

```python
X, y = make_classification(n_samples=200, n_features=2, n_classes=2, n_redundant=0, 
                           n_clusters_per_class=1, class_sep=0.5)

fig, ax = plt.subplots()
plt.scatter(X[:, 0], X[:, 1], c=[c_list[idx] for idx in y], cmap='coolwarm', alpha=0.9)
ax.set_title("Dataset", fontsize = 20, fontweight = "bold")
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

```python
tree = DecisionTreeClassifier()
tree.fit(X, y)

y_pred_train = tree.predict(X)
accuracy_train = accuracy_score(y, y_pred_train)
print(f"Accuracy on training set: {accuracy_train:.3f}")
```

```python
# Create a grid of points to evaluate the decision function
xx, yy = np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
Z = tree.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

fig, ax = plt.subplots()
plt.contourf(xx, yy, Z, alpha=0.5, cmap='coolwarm_r')
ax.set_title("Decision Boundary", fontsize = 20, fontweight = "bold")
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```

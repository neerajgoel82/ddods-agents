```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import seaborn as sns
sns.set()
plt.style.use(["seaborn-darkgrid"])
```

```python
df=pd.DataFrame({'x': np.random.normal(10, 1.2, 20000), 'y': np.random.normal(10, 1.2, 20000), 'group': np.repeat('A',20000) })
tmp1=pd.DataFrame({'x': np.random.normal(14.5, 1.2, 20000), 'y': np.random.normal(14.5, 1.2, 20000), 'group': np.repeat('B',20000) })
tmp2=pd.DataFrame({'x': np.random.normal(9.5, 1.5, 20000), 'y': np.random.normal(15.5, 1.5, 20000), 'group': np.repeat('C',20000) })
df=pd.concat((df, tmp1,tmp2))
df = df.reset_index(drop = True)

df.head()
```

```python
X = df[["x", "y"]].values
```

## Default Scatter Plot

```python
fig, ax = plt.subplots(figsize = (12, 9))
plt.scatter(X[:, 0], X[:, 1])
ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.tick_params(axis='both', which='major', pad=15)
ax.set_title("Default Scatter Plot", fontsize = 25, fontweight="bold")
plt.xlim(3, 20);
```

## Scatter Plot with\n Reduced Marker Size

```python
fig, ax = plt.subplots(figsize = (12, 9))
plt.scatter(X[:, 0], X[:, 1], s=0.9)
ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.tick_params(axis='both', which='major', pad=15)
ax.set_title("Scatter Plot with\n Reduced Marker Size", fontsize = 25, fontweight="bold")
plt.xlim(3, 20);
```

## Hexbin Plots

```python
ax = df.plot(kind = "hexbin",
        x = 'x',
        y = 'y', gridsize = 13, figsize=(12, 9),  sharex=False, cmap="Blues")

ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_title("Hexbin Plot", fontsize = 25, fontweight="bold")
ax.tick_params(axis='both', which='major', pad=15);
```

## KDE Plot

```python
fig, ax = plt.subplots(figsize = (12, 9))
sns.kdeplot(data = df, x="x", y="y", cmap="Blues", shade=True)
plt.xlim(4, 20)
plt.ylim(5, 22)
ax.set_xticklabels(range(4, 22, 2), fontsize = 15)
ax.set_yticklabels(range(6, 25, 2), fontsize = 15)
ax.set_xlabel(None)
ax.set_ylabel(None)
ax.set_title("KDE Plot", fontsize = 25, fontweight="bold")
ax.tick_params(axis='both', which='major', pad=15);
```

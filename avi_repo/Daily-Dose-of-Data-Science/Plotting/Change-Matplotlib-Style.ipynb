# Style Matplotlib Plots To Make Them More Attractive

Post Link: [Substack](https://avichawla.substack.com/p/style-matplotlib-plots-to-make-them)

LinkedIn Post: [LinkedIn](https://www.linkedin.com/feed/update/urn:li:activity:7031211065422004225/)

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=100, noise=0.05, random_state=42)
```

```python
plt.style.available
```

```python
fig = plt.figure(dpi=600, figsize=(10, 20), tight_layout=True)
available = ['default'] + plt.style.available
for i, style in enumerate(available):
    with plt.style.context(style):
        ax = fig.add_subplot(12, 4, i + 1)
        ax.scatter(X[y==0, 0], X[y==0, 1], c="#b3de69")
        ax.scatter(X[y==1, 0], X[y==1, 1], c="#80b1d3")
    ax.set_title(style)
```

# Visualizing Google Search Trends of 2022 using Python

Post Link: [Substack](https://avichawla.substack.com/p/visualizing-google-search-trends)

LinkedIn Post: [LinkedIn](https://www.linkedin.com/posts/activity-7016005911064252416-vzZz?utm_source=post_nba&utm_medium=member_desktop&utm_campaign=copy)

Repo Link: https://github.com/joweich/year-in-search-trends

```python
import yearinsearchtrends as yist
```

```python
keywords = ["Wordle", "Ukraine", "COVID", "IPL", "Twitter Takeover", 
            "Johnny Depp",  "JWST", "Stable Diffusion",
            "India Pakistan", "Queen Elizabeth", 'Roger Federer', "Rishi Sunak", "Tech Layoffs", "Iran",
            'FIFA World Cup' , "FTX", "ChatGPT",]
```

```python
df = yist.get_interest_over_time(
    keywords, "2022-01-01 2023-01-01"
)
yist.draw_ridgeplot(df)
```

# Hexbin Plots As A Richer Alternative to Scatter Plots

Post Link: [Substack](https://avichawla.substack.com/p/hexbin-plots-as-a-richer-alternative)

LinkedIn Post: [LinkedIn](https://www.linkedin.com/posts/avi-chawla_python-datascience-activity-6996424549546303488-inqc?utm_source=share&utm_medium=member_desktop)

```python
import pandas as pd
```

```python
df = pd.read_csv("../_Datasets/Hexbin_Plot_Example.csv")
```

## Scatter

```python
df.plot(kind='scatter',x='x', y='y')
```

## Hexbin

```python
df.plot(kind='hexbin',
        x='x',
        y='y', 
        gridsize = 13)
```

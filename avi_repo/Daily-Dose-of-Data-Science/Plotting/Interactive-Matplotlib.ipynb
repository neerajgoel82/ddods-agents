# Who Said Matplotlib Cannot Create Interactive Plots?

Post Link: [Substack](https://avichawla.substack.com/p/who-said-matplotlib-cannot-create)

LinkedIn Post: [LinkedIn](https://www.linkedin.com/feed/update/urn:li:ugcPost:7038823248310001664/)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import TextBox
```

```python
# enable interactive backend
%matplotlib widget

fig, ax = plt.subplots()
fig.subplots_adjust(bottom=0.2)

x = np.arange(-20, 20, 0.001)
line, = ax.plot(x, np.zeros_like(x), lw=2)

def submit(expression):      # function to update the plot
    ydata = eval(expression) # evaluate expression
    line.set_ydata(ydata)    # update plot
    ax.relim()               # recompute data limit
    ax.autoscale_view()
    plt.draw()

# add text box
text_box = fig.add_axes([0.1, 0.05, 0.8, 0.075])
text_box = TextBox(text_box, "Evaluate", textalignment="center")
text_box.on_submit(submit)

plt.show()
```

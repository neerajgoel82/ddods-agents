```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot(style):
    plt.style.use(style)

    x = np.linspace(0, 6, 20)
    df = pd.DataFrame({"x": x, 
                       "sin(x)": np.sin(x), 
                       "sin(2*x)": 3 + np.sin(2*x) , 
                       "sin(3*x)": 6 + np.sin(3*x)}
                     )
    fig, ax = plt.subplots()

    plt.plot("x", "sin(x)",   data=df, label = "sin(x)")
    plt.plot("x", "sin(2*x)", data=df, label = "sin(2x)")
    plt.plot("x", "sin(3*x)", data=df, label = "sin(3x)")
    plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
    
    ax.set_yticks([])
    plt.show()
```

```python
plot(["default"])
```

```python
plot(['ipynb'])
```

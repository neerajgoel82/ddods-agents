```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
plt.style.use(["seaborn-darkgrid"])
c_list = ['#46C7C7', '#00B9FF', '#FF7051'][::-1]
sns.set()
```

## Data

```python
r = np.linspace(0.3, 1, 30)
theta = np.linspace(0, 4*np.pi, 30)
x = r * np.sin(theta)
y = r * np.cos(theta)
```

## Default Ordering

```python
fig, ax = plt.subplots(1, figsize=(5, 5))

ax.plot(x, y, lw=5, color = c_list[0])
ax.scatter(x, y, s=200  , color = c_list[1])
ax.set_title('Lines on top', fontsize = 20, fontweight="bold")
ax.set_yticklabels([])
ax.set_xticklabels([]);
```

## Controlled Ordering

```python
fig, ax = plt.subplots(1, figsize=(6, 6))

ax.plot(x, y, lw=5,c = c_list[0],  zorder = 1)
ax.scatter(x, y, s=200,  c = c_list[1], zorder=2)  # move dots on top of line
ax.set_title('Dots on top', fontsize = 20, fontweight="bold")
ax.set_yticklabels([])
ax.set_xticklabels([]);
```

## Line Behind Grid

```python
fig, ax = plt.subplots(1, figsize=(6, 6))

ax.plot(x, y, lw=7,c = c_list[0],  zorder = 0)
ax.scatter(x, y, s=200,  c = c_list[1], zorder=2)  # move dots on top of line
ax.set_title('Line behind grid', fontsize = 20, fontweight="bold")
ax.set_yticklabels([])
ax.set_xticklabels([]);
```

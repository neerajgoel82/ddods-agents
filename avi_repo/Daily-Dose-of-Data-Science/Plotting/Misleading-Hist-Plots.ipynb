```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
plt.style.use(["seaborn-darkgrid"])
```

```python
df = pd.read_csv("../_Datasets/Price_data.csv")
df.head()
```

## Different Bins

```python
fig, ax = plt.subplots(3,3, figsize = (15, 9))

ax[0][0].hist(df.Price, bins = 10, color = "lightsalmon")
ax[0][1].hist(df.Price, bins = 15, color = "lightsalmon")
ax[0][2].hist(df.Price, bins = 20, color = "lightsalmon")
ax[1][0].hist(df.Price, bins = 25, color = "lightsalmon")
ax[1][1].hist(df.Price, bins = 30, color = "lightsalmon")
ax[1][2].hist(df.Price, bins = 35, color = "lightsalmon")
ax[2][0].hist(df.Price, bins = 40, color = "lightsalmon")
ax[2][1].hist(df.Price, bins = 45, color = "lightsalmon")
ax[2][2].hist(df.Price, bins = 50, color = "lightsalmon")

ax[0][0].set_xticklabels([])
ax[0][1].set_xticklabels([])
ax[0][2].set_xticklabels([])
ax[1][0].set_xticklabels([])
ax[1][1].set_xticklabels([])
ax[1][2].set_xticklabels([])

ax[2][0].set_xticks([75, 100, 125, 150, 175, 200])
ax[2][1].set_xticks([75, 100, 125, 150, 175, 200])
ax[2][2].set_xticks([75, 100, 125, 150, 175, 200])

ax[0][0].set_title("Bin Count = 10", fontweight = "bold", fontsize = 15)
ax[0][1].set_title("Bin Count = 15", fontweight = "bold", fontsize = 15)
ax[0][2].set_title("Bin Count = 20", fontweight = "bold", fontsize = 15)
ax[1][0].set_title("Bin Count = 25", fontweight = "bold", fontsize = 15)
ax[1][1].set_title("Bin Count = 30", fontweight = "bold", fontsize = 15)
ax[1][2].set_title("Bin Count = 35", fontweight = "bold", fontsize = 15)
ax[2][0].set_title("Bin Count = 40", fontweight = "bold", fontsize = 15)
ax[2][1].set_title("Bin Count = 45", fontweight = "bold", fontsize = 15)
ax[2][2].set_title("Bin Count = 50", fontweight = "bold", fontsize = 15)

ax[0][0].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[0][1].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[0][2].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[1][0].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[1][1].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[1][2].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[2][0].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[2][1].tick_params(axis='y', which='major', pad=10, labelsize = 13)
ax[2][2].tick_params(axis='y', which='major', pad=10, labelsize = 13)

ax[2][0].tick_params(axis='x', which='major', pad=10, labelsize = 13)
ax[2][1].tick_params(axis='x', which='major', pad=10, labelsize = 13)
ax[2][2].tick_params(axis='x', which='major', pad=10, labelsize = 13)

plt.suptitle("Same Data, VERY Different Histograms", fontweight = "bold", fontsize = 20)
plt.show();
```

## Violin Plot

```python
sns.violinplot(data = df.Price, color="lightsalmon")
plt.xlabel("Price")
plt.tick_params(axis='x', which='major', pad=10, labelsize = 13)
plt.title("Violin Plot", fontweight = "bold", fontsize = 20);
```

## KDE Plot

```python
sns.kdeplot(df.Price, color="lightsalmon", fill=True)
plt.xlabel("Price", labelpad=10)
plt.ylabel("Density", labelpad=10)
plt.tick_params(axis='x', which='major', pad=10, labelsize = 13)
plt.tick_params(axis='y', which='major', pad=10, labelsize = 13)
plt.title("KDE Plot", fontweight = "bold", fontsize = 20);
```

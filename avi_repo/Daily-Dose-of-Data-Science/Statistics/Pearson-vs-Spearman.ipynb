```python
# Compute Pearson correlation coefficient for linear data
linear_pearson, _ = pearsonr(x, linear_data)
print("Pearson correlation for linear data:", linear_pearson)

# Compute Spearman correlation coefficient for linear data
linear_spearman, _ = spearmanr(x, linear_data)
print("Spearman correlation for linear data:", linear_spearman)

# Compute Pearson correlation coefficient for polynomial data
poly_pearson, _ = pearsonr(x, poly_data)
print("Pearson correlation for polynomial data:", poly_pearson)

# Compute Spearman correlation coefficient for polynomial data
poly_spearman, _ = spearmanr(x, poly_data)
print("Spearman correlation for polynomial data:", poly_spearman)
```

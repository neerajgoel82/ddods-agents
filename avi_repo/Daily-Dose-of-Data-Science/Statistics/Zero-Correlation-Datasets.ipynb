# Be Cautious Before Drawing Any Conclusions Using Summary Statistics

LinkedIn Post: [LinkedIn](https://www.linkedin.com/feed/update/urn:li:activity:7053281636926275584/)

```python
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
from sklearn.datasets import make_circles, make_blobs

sns.set()

plt.rcParams['font.family'] = ['Comic Sans MS', 'sans-serif']
warnings.filterwarnings("ignore")
np.random.seed(42)
```

```python
color = '#2471A3'
```

```python
fig, ax = plt.subplots(3,3, figsize = (16, 12))

X = np.linspace(0, 3*np.pi, num=100)
Y = np.sin(X) + np.random.normal(0, 0.1, size=X.shape)
sns.scatterplot(X, Y, ax = ax[0][0], s = 25, color = color )

X = np.linspace(-5, 5, num=100)
Y = X**2 + np.random.normal(0, 1, size=X.shape)
sns.scatterplot(X, Y, ax = ax[0][1], s = 25, color = color)

X, y = make_circles(n_samples=200, noise=0.04, random_state=42, factor=0.5)
sns.scatterplot(X[:, 0], X[:, 1], ax = ax[0][2], s = 25, color = color)

X_inner = X[y == 1]
sns.scatterplot(X_inner[:, 0], X_inner[:, 1], ax = ax[2][0], s = 25, color = color)


X = np.linspace(1, 10, num=100)
Y = np.log(X) + np.random.normal(0, 0.05, size=X.shape)
X = np.concatenate((-X[::-1], X))
Y = np.concatenate((Y[::-1], Y))
sns.scatterplot(X, Y, ax = ax[1][1], s = 25, color = color)


X = np.linspace(-3, 3, num=100)
Y = np.exp(-X**2) + np.random.normal(0, 0.03, len(X))
sns.scatterplot(X, Y, ax = ax[1][2], s = 25, color = color)


# Generate a dataset with four blobs located at (+-2,+-2)
centers = [(2, 2), (-2, -2), (-2, 2), (2, -2)]
X, y = make_blobs(n_samples=200, centers=centers, cluster_std=0.3)
sns.scatterplot(X[:, 0], X[:, 1], ax = ax[1][0], s = 25, color = color)

X = np.linspace(0, 4, num=100)
Y = np.concatenate([np.zeros(20), np.ones(20), np.zeros(20), np.ones(20), np.zeros(20)]) + np.random.normal(0, 0.05, size=X.shape)
sns.scatterplot(X, Y, ax = ax[2][1], s = 25, color = color)

# Generate a dataset with an absolute value relationship
X = np.linspace(-1, 1, num=100)
Y = np.abs(X + np.random.normal(0, 0.03, size=X.shape)) 
sns.scatterplot(X, Y, ax = ax[2][2], s = 25, color = color)


ax[0][0].set_xticklabels([])
ax[0][1].set_xticklabels([])
ax[0][2].set_xticklabels([])
ax[1][0].set_xticklabels([])
ax[1][1].set_xticklabels([])
ax[1][2].set_xticklabels([])
ax[2][0].set_xticklabels([])
ax[2][1].set_xticklabels([])
ax[2][2].set_xticklabels([])

ax[0][0].set_yticklabels([])
ax[0][1].set_yticklabels([])
ax[0][2].set_yticklabels([])
ax[1][0].set_yticklabels([])
ax[1][1].set_yticklabels([])
ax[1][2].set_yticklabels([])
ax[2][0].set_yticklabels([])
ax[2][1].set_yticklabels([])
ax[2][2].set_yticklabels([])

ax[0][0].set_title("Sinusoid", fontsize = 17, fontweight = "bold")
ax[0][1].set_title("Quadratic", fontsize = 17, fontweight = "bold")
ax[0][2].set_title("Concentric Circles", fontsize = 17, fontweight = "bold")
ax[1][0].set_title("Blobs", fontsize = 17, fontweight = "bold")
ax[1][1].set_title("Symmetic Log", fontsize = 17, fontweight = "bold")
ax[1][2].set_title("Bell Curve", fontsize = 17, fontweight = "bold")
ax[2][0].set_title("Circle", fontsize = 17, fontweight = "bold")
ax[2][1].set_title("Step function", fontsize = 17, fontweight = "bold")
ax[2][2].set_title("Absolute value", fontsize = 17, fontweight = "bold")

plt.show();
```
